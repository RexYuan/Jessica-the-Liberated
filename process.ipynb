```python
connection = pymysql.connect(host='localhost',
                             user='root',
                             database='census99')

# SQL command to count rows in the "people" table
count_command = "SELECT COUNT(*) FROM people"

# Create a cursor object to interact with the database
with connection.cursor() as cursor:
    # Execute the SQL query
    cursor.execute(count_command)

    # Fetch the result (only one row, one column)
    result = cursor.fetchone()

    # Print the result; result[0] contains the count
    print(f"Total rows in the people table: {result[0]}")
```
